## Image Scraper

This script was used to scrape images from the web.

```python
# Importing the library
from google_images_download import google_images_download   

# Class instantiation
response = google_images_download.googleimagesdownload()    

# Create list of arguments. I modify this list based on what I want to collect.
arguments = {"keywords":"c7 silver corvette,c7 yellow corvette,c6 yellow corvette,c6 silver corvette,c5 yellow corvette,c5 silver corvette,c4 yellow corvette,c4 silver corvette,c3 white corvette,c2 white corvette classic,c2 red corvette classic,c1 white corvette classic,c1 red corvette classic",
             "limit":100,"print_urls":False}   

# Passing the arguments to the function
paths = response.download(arguments)

# Printing absolute paths of the downloaded images
print(paths)                           
```
